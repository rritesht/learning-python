# Unwanted Outliers

@ Box Plot


        @ MEDIAN : It will seperate the data into two Equal Parts.
        
        ------1-------Q1--------2--------Q2---------3--------Q3------4-------   -> Line
        
   QUESTION: If you want to deivide into 4 equal parts how many placeses ou will cut the above line?
   
   
      
      Q1, Q2, Q3  
      
      Q1 = 1 X ((n-1)/4 +1)
      
      Q2 = 2 X ((n-1)/4 +1)
      
      Q3 = 3 X ((n-1)/4 +1)
      
      IQR  = Q3 -Q1
      
      
      Upper limit = Median + 1.5 X IQR
      
      Lower limit = Median - 1.5 X IQR
      
      
   @ Note: The Observations which are above Upper limit and Below Lower limit are considered as Outliers.

@Outliers: 
    
    However, you need justification to remove the data.
    
    You should never remove an outlier just because it's a "big_number"
    
    the Big  number could be very informative to our model some times.

@ Good Reasons:
    
    1. Suspicious measurments that are unlikely to be real data. 
    
            Ex: 1750 year transaction  all other are 1990 above
            
    2. Outliers that belong in a different population.
    
    3. Outliers that belong to a different problem.

# Missing Categorical Data

```python
df.head(20)
```

@ Label missing categorical Data:

    @ we can't ignore simply missing values in our dataset.
    
    @ we need to handle missing values for very practical reason , SKLEARN doesn't accept those.
    
    
 @ 2 most common ways to deal with missing values :
    
   1. Dropping  : We Will drop the observations with missing values
            
   2. Imputing :  The missing values are replaced based on the values from various obsertvations.
   
   
   
  $ Why dropping missing values not work :
  
      * When we are dropping the observations, it means your dropping the information.
      
      * The fact that the values was missing may be informative itself.
      
      * In the real world, we often need to make predictions on new data even if some of the features are missing ...!
      
      
      
   $ Why Imputing missing values not work(Sucks the model):
   
       * The values was originally missing but we are filling it, which always leads to a loss in the information. 
       
       * Missingness is almost always informative in itself, and we should tell to our algorithm if a value was missing.
       
       * Even if you buid a model to impute your values, you are not adding any real infoarmation. you are reinforcing the patterns already provided by the other features.
