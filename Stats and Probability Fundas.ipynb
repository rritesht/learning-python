# Influence of Outliers

    Shiva    Sai     Manoj   Murali   kalyan

    33        44      99       56      53
    
    
    
    Average marks = 285/5 = 57

 1. Is manoj marks are properly collected or any mistake happend?
 2. what is the reason behind manoj marks?   (Inferential Statistics)

    Re calculating average marks by excluding Manoj marks
    
    285 - 99 = 186/4= 46.5

From the above understanding we can easily understand outlier skews the mean completely.

```python
import numpy as np 

mark = [33,44,99,56,53]

marks = np.array(mark)

avg_marks = np.mean(marks)

print("Average/Mean before removing the outlier marks :",avg_marks)

marks_otlier_removed = np.delete(marks,2)

avg_rem_otl = np.mean(marks_otlier_removed)

print("Average/Mean after removing outlier marks : ",avg_rem_otl)

```

```python
from matplotlib import pyplot as plt

plt.plot(marks)

plt.show()
```

```python
# median

print("Median before removing the outlier marks",np.median(marks))

print("Median after removing outlier marks",np.median(marks_otlier_removed))
```

# IQR -{Interquartile Range}

## Measures of spread/Dispersion

    1. Range : Difference between the largest and smallest values (0 or > 0 non negetive)
    2. Quartile 1: The middle number between the smallest number and median.
    3. Quartile 2: The middle number between the largest number and median.
    4. Interquartile range:  Difference between Q3 and Q1
    5. Varaince and Standard deviation: A measure of variation in data.

   Shiva    Sai     Manoj   Murali   kalyan  mythili  rishi

## Characterstics and limitations of Range

    1. Computed using 2 data points.
    2. Outliers in data may give a false indication of how spread out it is.
    3. Useful if data is closely clustered and there are many points.
    4. To eliminate outliers interquartiles range is often considered.
