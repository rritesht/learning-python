#### What is linear regression?
    
    In statistics, linear regression is a approach to identify and modeling the relationship between a dependent and one or more independent variables.
    
    y = mx + c {m = slop, c = y- intercept, X_ intercept = -c/m}
    
    1. The best fit line which is very minimal distance from all points or minimum error 
    2. Loss Function (find out the most suitable or nearer values of "m" and "c")
