```python
import numpy as np 

import sklearn

from sklearn import model_selection
```

```python
# creating an array matrix
data_matrix = df1.values

# Extracted all input variables into one variable
X = data_matrix[:,0:4]

# Extracted all target variables into one variable
Y = data_matrix[:,4]

# Defining training and validation data portions
validation_size = 0.20

# Fixing the randomness
seed = 7

# Splitiing data into training and testing with help of model selection
X_train,X_validation,Y_train,Y_validation=model_selection.train_test_split(X,Y, test_size= validation_size, random_state = seed)
```

```python
from sklearn.metrics import accuracy_score

from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report

```

```python
# Evalution metric

scoring = 'accuracy'
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
from sklearn.linear_model
```

```python
# Algorithms

models = []

models.append(('LDA',LinearDiscriminantAnalysis()))



# .................................
#..................................

#..................................
#.................................


```

```python
models
```

```python
# evalution for each model with process specified

results = []

names = []

for name, model in models:
    kfold = model_selection.KFold(n_splits=10,random_state=seed)
    cv_results= model_selection.cross_val_score(model,X_train,Y_train,cv=kfold,scoring=scoring)
    results.append(cv_results)
    names.append(name)
    display_msz = "%s: %f (%f)" %(name, cv_results.mean(),cv_results.std())
    print(display_msz)
```

# Predictions

```python
# making predictions

lda = LinearDiscriminantAnalysis()

lda.fit(X_train,Y_train)

predictrions = lda.predict(X_validation)
```

```python
accuracy_score(Y_validation,predictrions)
```

```python
confusion_matrix(Y_validation,predictrions)
```

```python
print(classification_report(Y_validation,predictrions))
```
