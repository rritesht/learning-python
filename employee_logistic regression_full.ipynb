```python
# Confusion Matrix

                            Actual

                     +ve               -ve
        
        
       +ve          {TP}               {FP}      total Predicted +ves
    
pred   

       -ve          {FN}               {TN}      total predicted -ves
    
            Actual total +ves     Actaula total -ves

    

Precision = TP/(TP +FP)   {correct positive predictions / Total positive predictions}

Specificity/True Negetive rate

# precision always on predictions

Recall = TP/(TP + FN)     {correct positive predictions/ Total Actual Positives}

Sensitivity/True positive rate
    
    
# Accuaracy

Accuracy = TP +TN /(TP +TN +FP+FN)


# F1 Score

F1 = 2 * Precision * Recall/(Precision + Recal)
```

# Regularized Logistic regression

"Regularization is a technique used to prevent overfitting by artificially penalizing model coefficients"

It adds a penalty factor to the cost function.

It will dampens coffeficients or remove the features entirly, depending on the type of panalty used.

The strength of that penalty factor is tunable.


@ types of Penalties:

    1. L1 Penalizes the absolute size of model cofficients
    
    2. L2 Penalizes the squared size of the model coefficients.




@ Helper function:

    1. Fits any classification model 
    2. makes predictions
    3. Extracts the positive probabilities
    4. Plots them



```python
def fit_and_plot_classifier(clf_algo):
    
    # Fit the model
    
    clf_algo.fit(X,y)
    
   
    # Predict and take seocend value of each prediction
    
    pred = clf_algo.predict_proba(X)
    
    pred = [p[1] for p in pred]
    
    
    # Plot
    
    plt.scatter(X,y)
    
    plt.plot(X,pred,"--")
    
    plt.show()
    
    
    # Return fitted model and predictions
    
    return clf_algo,pred
    
    
```

```python
# logistic regression

clf_algo,pred = fit_and_plot_classifier(LogisticRegression())
```

### building model pipelines

@ Creating a single pipeline dictionary with pipelines for each algorithm.

    @ Keys:
        
        1. "l1" for L1 - regularized logistic regression
        
        2. "l2" for L2 - regularized logistic regression
        
        3. "rf" for random forest
        
        4. "gb" for gradient boosted tree
        
  
  
  @ Each pipeline should standardize the data first
  
  @ random_state = 123 for each algorithm to ensure replicable results.



@  StandardScalar()

    It will bring all the features into the same scale.
    
@ random_state: 

    it will extract the same random events based on our declaration.
    
@


```python
# pipeline dictionary

pipelines = {
    "l1":make_pipeline(StandardScaler(),LogisticRegression(penalty='l1',random_state= 123)),
    
    "l2":make_pipeline(StandardScaler(),LogisticRegression(penalty='l2',random_state= 123)),
    
    "rf":make_pipeline(StandardScaler(),RandomForestClassifier(random_state= 123)),
   
    "gb":make_pipeline(StandardScaler(),GradientBoostingClassifier(random_state= 123))
    
    
}
```

@ There are two types of parameters we need to take care when implementing MLalgo's

@ Model Parameters:
    These can be learn directly from the training data

@ Hyperparameters:
    we need to tune.
    
    

```python
# List of tunable hyperperameters of our Logistic Pipeline

pipelines['l1'].get_params()
```

```python
# logistic regression hyperperameters

l1_hyperperameters = {'logisticregression__C': np.linspace(1e-3, 1e3,10)}

l2_hyperperameters = {'logisticregression__C': np.linspace(1e-3, 1e3,10)}
```

```python
# List of tunable hyperperameters of our Logistic Pipeline

pipelines['rf'].get_params()
```

```python
# Random Forest Hyperperameters

rf_hyperperameters = {'randomforestclassifier__n_estimators':[100, 200],
                     'randomforestclassifier__max_features':['auto','sqrt',0.33]}
```

```python
pipelines['gb'].get_params()
```

```python
# Hyperperameters for Gradient boosted trees



gb_hyperperameters = {'gradientboostingclassifier__n_estimators':[100, 200],
                     'gradientboostingclassifier__learning_rate':[0.05,0.1,0.2],
                     'gradientboostingclassifier__max_depth':[1,3,5]}
```

```python
np.linspace(1e-3, 1e3,10)
```

```python
# creating hyperperamenters dictionary

hyperperameters = {"l1":l1_hyperperameters,
                  "l2":l2_hyperperameters,
                  'rf': rf_hyperperameters,
                  'gb': gb_hyperperameters}
```

```python
# creating empty dictionary  called fitted_models

fitted_models = {}

# Loop through the model pipelines, tuning each one and saving it to fit models

for name, pipe in pipelines.items():
    # creating  cross_validataino object from pipeline and hyperperameters
    
    model = GridSearchCV(pipe, hyperperameters[name],cv = 10,n_jobs = -1)
    
    # fitting model on X_train and y_train
    
    model.fit(x_train,y_train)
    
    # storing fitted models in the fitted_library
    
    fitted_models[name]= model
    
    
    print(name, "Fitting completed")
```
