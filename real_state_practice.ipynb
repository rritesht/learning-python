## Ridge Regression

```python
df_1 = pd.read_csv('finaldf.csv')
```

```python

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
Y = df_1[['tx_price']]
X = df_1.drop(['tx_price'],axis =1)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.30,random_state =7)
```

```python
rig = Ridge(alpha=1)
rig.fit(X_train,Y_train)
predict = rig.predict(X_test)
r2_score(Y_test,predict)
```

## Lasso Regression

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=1000)
lasso.fit(X_train,Y_train)
prid = lasso.predict(X_test)
r2_score(Y_test,prid)
```

## Elastic Net

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet(alpha=0.1)
elastic.fit(X_train,Y_train)
pre = elastic.predict(X_test)

print('Coefficients: \n', elastic.coef_)
print(r2_score(Y_test,pre))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
pipelines = {'Ridge':make_pipeline(StandardScaler(),Ridge()),
             'Lasso':make_pipeline(StandardScaler(),Lasso()),
            'Elasticnet':make_pipeline(StandardScaler(),ElasticNet())}
```

```python
pipelines['Elasticnet']
```

```python
pipelines['Elasticnet'].get_params()
```

```python
# Hyperparameter
Lasso_Hyperparameter = {'lasso__alpha':[0.001,0.005,0.01,0.1,0.5,1,1.5,2,10]
  }
Ridge_Hyperparameter = {'ridge__alpha':[0.001,0.005,0.01,0.1,0.5,1,1.5,2,10]
    }
Elasticnet_Hyperparameter = {'elasticnet__alpha':[0.001,0.005,0.01,0.1,0.5,1,1.5,2,10]
    }
```

```python
hyperperameters = {'Lasso':Lasso_Hyperparameter,
                 'Ridge' : Ridge_Hyperparameter,
                 'Elasticnet':Elasticnet_Hyperparameter}
```

```python
from sklearn.model_selection import GridSearchCV
# creating empty dictionary  called fitted_models

fitted_models = {}

# Loop through the model pipelines, tuning each one and saving it to fit models

for name, pipe in pipelines.items():
    # creating  cross_validataino object from pipeline and hyperperameters
    
    model = GridSearchCV(pipe, hyperperameters[name],cv = 10)
    
    # fitting model on X_train and y_train
    
    model.fit(X_train,Y_train)
   
    fitted_models[name]= model 
    
    # storing fitted models in the fitted_library

    print(name, "Fitting completed")
    


    
    
```

```python
for name, model in fitted_models.items():
    print(name, model.best_score_)
```
