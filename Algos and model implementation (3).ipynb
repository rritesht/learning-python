```python
# Function for creating pipelines

from sklearn.pipeline import make_pipeline

# For Standardization

from sklearn.preprocessing import StandardScaler

# Importing Elastic Net, Ridge regression, and Lasso
from sklearn.linear_model import ElasticNet,Ridge, Lasso

# Importing random forest and Gradiant boosted trees

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
make_pipeline(StandardScaler(),Lasso(random_state = 123))
```

```python
# creating pipeline dictionary

pipelines = {'Lasso': make_pipeline(StandardScaler(),Lasso(random_state = 123)),
            'ridge': make_pipeline(StandardScaler(),Ridge(random_state = 123))}
```

```python
#create a pipeline for Elastic Net
pipelines['Enet'] = make_pipeline(StandardScaler(),ElasticNet(random_state=123))

#create a pipeline for Elastic Net
pipelines['rf'] = make_pipeline(StandardScaler(),RandomForestRegressor(random_state=123))

#create a pipeline for Elastic Net
pipelines['gb'] = make_pipeline(StandardScaler(),GradientBoostingRegressor(random_state=123))
```

```python
# checking the pipeline for algorithms are added or net

for key, value in pipelines.items():
    print(key,type(value))
```
